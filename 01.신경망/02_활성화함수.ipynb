## 활성화 함수

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.linspace(-5,5,501)
```

- heaviside step function

```python
def heaviside(x):
  y = np.zeros_like(x)
  y[x > 0] = 1            # x <= 0 이면 0, x > 1 이면 1
  return y
```

```python
plt.plot(xs, heaviside(xs))
plt.grid()
plt.show()
```

- Sigmoid

```python
def sigmoid(x):
  return 1/ (1 + np.exp(-x))
```

```python
plt.plot(xs, sigmoid(xs))
plt.grid()
plt.show()
```

- hyperbolic tangent

```python
plt.plot(xs, np.tanh(xs))
plt.grid()
plt.show()
```

- relu(rectified linear unit)

```python
def relu(x):
  return np.maximum(0, x)
```

```python
plt.plot(xs, relu(xs))
plt.grid()
plt.show()
```

- leaky relu

```python
def leaky_relu(x):
  return np.maximum(0.01*x, x)
```

```python
plt.plot(xs, leaky_relu(xs))
plt.grid()
plt.show()
```

- softplus

```python
def softplus(x):
  return np.log(1 + np.exp(x))
```

```python
plt.plot(xs, softplus(xs))
plt.grid()
plt.show()
```

- softmax

```python
def softmax(x):
  x = x - np.max(x)          # overflow 방지
  return np.exp(x) / np.sum(np.exp(x))
```

```python
yhat = np.array([-1,1,3])
softmax(yhat)
```

```python
np.sum(softmax(yhat))
```
